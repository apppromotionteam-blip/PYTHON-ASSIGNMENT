**Python Basics Questions**

1. What is Python, and why is it popular ?
    -Python is a popular programming language known for its simple syntax, easy learning curve, and wide range of applications such as web development, data science, AI, automation, and software development. It is popular because it is easy to read, has many libraries, works on multiple platforms, and has a large developer community.
2.  What is an interpreter in Python?
    -A Python interpreter is a program that reads and executes Python code line by line, converting it into instructions the computer can understand and run.
3.  What are pre-defined keywords in Python?
    -Pre-defined keywords are reserved words in Python that have special meanings and cannot be used as variable, function, or class names. Examples include if, else, for, while, def, class, import, return, and True.

4. Can keywords be used as variable names?

    -No, Python keywords cannot be used as variable names because they are reserved words with special meanings.

5. What is mutability in Python?

    -Mutability is the ability of an object to be changed after it is created.

6. Why are lists mutable, but tuples are immutable?

    -Lists are mutable, meaning their elements can be added, removed, or modified. Tuples are immutable, meaning their contents cannot be changed after creation.

7. What is the difference between == and is operators in Python?

    -(==) checks whether two objects have the same value.
    is checks whether two variables refer to the same object in memory.

8. What are logical operators in Python?

    -Logical operators are used to combine conditional statements. They are: and,or,not
9. What is type casting in Python?

    -Type casting is the process of converting a value from one data type to another.

10. What is the difference between implicit and explicit type casting?

    =Implicit type casting: Python automatically converts one data type to another.
    Explicit type casting: The programmer manually converts the data type using functions like int(), float(), or str().

11. What is the purpose of conditional statements in Python?

    =Conditional statements allow a program to make decisions and execute different code blocks based on specified conditions.

12. How does the elif statement work?

    =The elif statement checks another condition if the previous if condition is false. It helps handle multiple conditions efficiently.

13. What is the difference between for and while loops?

    =A for loop is used when the number of iterations is known.
    A while loop is used when the loop should continue until a condition becomes false.

14. Describe a scenario where a while loop is more suitable than a for loop.

    A while loop is more suitable when the number of iterations is unknown, such as repeatedly asking a user for input until they enter a valid value.

4. Check if a Word is a Python Keyword

```python
import keyword

word = input("Enter a word: ")

if keyword.iskeyword(word):
    print(f"{word} is a Python keyword.")
else:
    print(f"{word} is not a Python keyword.")
```

5. List vs Tuple

```python
my_list = [10, 20, 30]
my_tuple = (10, 20, 30)

my_list[0] = 100
print("Modified List:", my_list)

try:
    my_tuple[0] = 100
except TypeError as e:
    print("Error:", e)

print("Tuple:", my_tuple)
```

6. Mutable and Immutable Arguments

```python
def modify(number, my_list):
    number += 10
    my_list.append(4)

x = 5
y = [1, 2, 3]

modify(x, y)

print("Integer:", x)
print("List:", y)
```

10. Type Casting with List Elements

```python
str_list = ["1", "2", "3", "4"]

int_list = [int(x) for x in str_list]

print("Original List:", str_list)
print("Integer List:", int_list)
```

14. Reverse a String Using While Loop

```python
text = input("Enter a string: ")

i = len(text) - 1

while i >= 0:
    print(text[i], end="")
    i -= 1
```
